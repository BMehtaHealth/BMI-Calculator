```python
def calculate_bmi(w, h):

    """calculating bmi of the patient"""
    if w <= 0 or h <= 0:
        return "Error: Weight and height must be positive numbers"
        """checking for valid input"""
    bmi = w / (h**2)
    return bmi

def category(bmi):
    """ categorising according to WHO """
    if bmi < 16 :
        return "Severly Underweight"
    elif 16 <= bmi < 18.5 :
        return "Underweight"
    elif 18.5 <= bmi < 25 :
        return "Normal"
    elif 25 <= bmi < 30 :
        return "Overweight"
    elif 30 <= bmi < 35 :
        return "Moderately Obese"
    elif 35 <= bmi < 40 :
        return "Severly Obese"
    elif  bmi > 40 :
        return "Morbidly Obese"
```

```python
calculate_bmi(100, 1.8)
```
